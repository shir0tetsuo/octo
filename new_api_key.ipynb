```python
# Generate a new DATABASE API Key with the private key
import engine.security
from pathlib import Path
import os

tokendata = ['WRITE_OK', 'READ_OK', os.getlogin()]  # This could be anything

print(
    engine.security.create_api_key(
        *tokendata, key_storage_file=Path('engine/key.json')
    ).decode()
)
```

```python
# Generate a new FRONTEND API Key with the private key
import engine.security
from pathlib import Path
import os

tokendata = ['user:303309686264954881','isLevel1']  # isLevel1 means basic user

APIKey=(
    engine.security.create_api_key(
        *tokendata, key_storage_file=Path('engine/key.json')
    ).decode()
)

engine.security.decrypt_api_key(APIKey, key_storage_file=Path('engine/key.json'))
```
