```python
kv = {'1': '1', '2': '2'}

ents = ['CREATE TABLE IF NOT EXISTS entities (']
for k, v in kv.items():
    indent = ' '*8
    ents.append(f'{indent}{k} {v}')
ents.append(')')
print('\n'.join(ents))
```

```python
def unwrap_schema(kv:dict, table_name:str):
    indent = ' '*8
    sqlcommand = [f'CREATE TABLE IF NOT EXISTS {table_name} ('] + [f'{indent}{k} {v}' for k,v in kv.items()] + [')']
    return '\n'.join(sqlcommand)
```
